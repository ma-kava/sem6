```python
import numpy as np
import matplotlib.pyplot as plt
from tempwages import *
```

# Load the data

```python
data_wages = np.loadtxt("mzdy.txt")
data_temps = np.loadtxt("teplota.txt")
```

# Prepare the data

```python
M = np.array(data_wages[:,1])
t_wages = data_wages[:,0]

T = np.array(data_temps[:,1])
t_temps = np.array(data_temps[:,0])
```

# Compute linear regression

```python
x = fit_wages(t_wages, M)
```

```python
n_points = 200

start = np.min(t_wages)
stop = np.max(t_wages)

ts_wages = np.linspace(start, stop, n_points)
y = x[0] + x[1] * ts_wages
plt.plot(ts_wages, y, 'red', label='data_wages')
plt.plot(t_wages, M, 'x')
```

```python
omega = 2 * np.pi / 365
x_temps = fit_temps(t_temps, T, omega)
print(x_temps)
```

```python
n_points = 200

start_temps = np.min(t_temps)
stop_temps = np.max(t_temps)

ts = np.linspace(start_temps, stop_temps, n_points)
y_temps = x_temps[0] + x_temps[1]*ts + x_temps[2]*np.sin(omega*ts) + x_temps[3]*np.cos(omega*ts)
plt.plot(ts, y_temps, 'red', label='data')
plt.plot(t_temps, T, 'x')
```
